**DATA EXPLORATION**



```python
from sklearn.feature_selection import SelectKBest
bestfeatures = SelectKBest(k = 9)

fit = bestfeatures.fit(x,y)

hiiscores = pd.DataFrame(fit.scores_)

hiicolumns = pd.DataFrame(x.columns)

hii = pd.concat([hiicolumns,hiiscores],axis = 1)
hii.columns = ['plc','Score']

print(hii)
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.25, random_state = 0)

```

```python
y_train.value_counts()
```

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
x_train = sc.fit_transform(x_train)
x_test = sc.transform(x_test)
```

```python

from sklearn import tree
# from sklearn.tree import DecisionTreeClassifier

dt = tree.DecisionTreeClassifier(max_depth = 4, criterion = 'entropy')
dt.fit(x_train,y_train)

y_pred = dt.predict(x_test)
y_pred
```

```python
from sklearn.metrics import ConfusionMatrixDisplay
ConfusionMatrixDisplay.from_predictions(y_test,y_pred)
```

**MODEL EVALUATION**

```python
testing_score = dt.score(x_test, y_test)
testing_score
```

```python
from sklearn.metrics import confusion_matrix
con_mat = confusion_matrix(y_test, y_pred)
con_mat
```

```python
from sklearn.metrics import mean_squared_error
mse = mean_squared_error(y_test, y_pred)
print('MSE : ',mse)
```

```python
from sklearn.metrics import precision_score
precision = precision_score(y_test, y_pred, average = 'binary')
print('Precision : %.3f' %precision)
```

```python
from sklearn.metrics import recall_score
recall = recall_score(y_test, y_pred, average = 'binary')
print('Recall : %.3f' %recall)
```

```python
from sklearn.metrics import f1_score
score = f1_score(y_test, y_pred, average = 'binary')
print('F1 Score : %.3f' %score)
```

```python
x = df.iloc[:,:17]
y = df.iloc[:,-1]

x.head()
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size = 0.30,random_state = 0)
```

```python
x_train.head()
```

```python
y_train.head()
```

**DATA EXPLORATION**

```python
from sklearn.preprocessing import StandardScaler

sd = StandardScaler()

sd.fit(x_train)

x_train = sd.transform(x_train)

x_test = sd.fit_transform(x_test)
```

```python
print(x_train.shape)

print(y_train.shape)
```

```python
from sklearn.neighbors import KNeighborsClassifier
model = KNeighborsClassifier(n_neighbors = 10)
model.fit(x_train,y_train)
```

```python
y_pred = model.predict(x_test)
y_pred
```

```python
y_test
```

**MODEL EVALUATION**

```python
from sklearn.metrics import accuracy_score
score = accuracy_score(y_test,y_pred)
score
```

```python
from sklearn.metrics import confusion_matrix

cm = confusion_matrix(y_test,y_pred)

cm
```

```python
from sklearn.metrics import mean_squared_error

mse = mean_squared_error(y_test,y_pred)

print("MSC: ",mse)
```

```python
from sklearn.metrics import precision_score

#calculate prediction

precision = precision_score(y_test,y_pred,average = 'binary')

print('Precision : %.3f' % precision)
```

```python
precision = precision_score(y_test,y_pred, labels = [0,1] , average = 'micro')

print('Precision : %.3f' % precision)
```

```python
from sklearn.metrics import recall_score

recall = recall_score(y_test,y_pred,average = 'binary')

print('Recall : %.3f' % recall)
```
